```python
from scipy import stats
import numpy as np
```

```python
df['value'].quantile(0.25)
#25th percentile of the data
```

```python
df['value'].quantile(0.75)
#75th percentile
```

```python
Q1=df['value'].quantile(0.25)
Q3=df['value'].quantile(0.75)
IQR=Q3-Q1
#interquantile range, which is 75th percentile - 25th percentile
```

```python
IQR
```

```python
lower_lim=Q1-1.5*IQR
upper_lim=Q3+1.5*IQR
#lower limit or lower fence of the data
#upper limit or upper fence of the data
```

```python
lower_lim
```

```python
upper_lim
```

```python
#left_outlier=df['value']<lower_lim
#right_outlier=df['value']>upper_lim
```

```python
outliers = []
for i in df['value']:
    if (i<lower_lim) or (i>upper_lim):
        outliers.append(i)
#detecting outlier in the dataset . we have set a condition that if value is lower than lower limit and if value is higher than upper limit we will treat it as outlier.
```

```python
len(outliers)
#these are the outliers in our dataset
```
